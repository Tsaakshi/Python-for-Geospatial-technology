Q.13. CALCULATE TOTAL NUMBER OF MISSING VALUES PRESENT IN DATA

```python
val1=data.isnull().sum()
mis_val=val1.sum()
print(f"Total number of missing values present in the data:", mis_val)

```

Q.14.CALCULATE PERCENTAGE OF MISSING VALUES PRESENT IN EACH COLUMN

```python
percent=(data.isnull().sum()*100/len(data)).round(2)
colm_val=pd.DataFrame({'passenger_type':data.columns,'percent': percent})
colm_val.sort_values(by = 'percent', inplace=True)
print(colm_val)
```
