```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("nba.csv")
```

```python
dataset.head(5)
```

```python
x=dataset.head(5)['Age']
y1=dataset.head(5)['Weight']
y2=dataset.head(5)['Height']
y3=dataset.head(5)['Salary']
plt.subplot(1,3,1)
plt.plot(x,y1,label="Line 1")
plt.title("Line Chart 1")
plt.subplot(1,3,2)
plt.plot(x,y2,label="Line 1")
plt.title("Line Chart 2")
plt.subplot(1,3,3)
plt.plot(x,y3,label="Line 1")
plt.title("Line Chart 3")

```
